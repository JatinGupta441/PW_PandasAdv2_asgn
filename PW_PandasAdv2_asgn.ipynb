```python
#Q3
new_df=df.reindex([3,0,1,2])
print(new_df.loc[2],"\n")
print(new_df.iloc[2])
```

```python
#In the output, new_df.loc[2] retrieves the row labeled 2, which contains the course name 'Big Data' and duration 6.
#On the other hand, new_df.iloc[2] retrieves the row at index position 2, which corresponds to the course name 
#'Machine Learning' and duration 3.
```

```python
#Q9
file_path=input("Enter a csv file path: ")
df2=pd.read_csv(file_path)
print(df2,"\n")
cat_col=input("Enter which column you want to convert to categorical data type: ")
print(df2[cat_col],"\n")
cat_ord=input("Enter in which order you want to sort: ")
cat_ord1=cat_ord.split(",")
df2[cat_col]=pd.Categorical(df2[cat_col],categories=cat_ord1,ordered=True)
print(df2[cat_col],"\n")
print(df2.sort_values(by=cat_col))
```
